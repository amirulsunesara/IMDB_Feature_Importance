```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

dataset_akas = pd.read_csv("title.akas.csv")
dataset_akas
```

```python
titles = dataset_akas.filter(['titleId','title','region','language'], axis=1)
titles=titles.loc[titles['region']=='US']
titles.shape
```

```python
dataset_basics = pd.read_csv("title.basics.csv")
dataset_basics.shape
```

```python
dataset_basics=(dataset_basics[dataset_basics['startYear']!="\\N"])
dataset_basics.rename(columns={'tconst':'titleId'}, inplace=True)
dataset_basics=dataset_basics.loc[(dataset_basics['titleType']=='movie')]
dataset_basics.shape
```

```python
titles= titles[titles.titleId.isin(dataset_basics['titleId'])]
```

```python
titles
```

```python
title_movies_region=pd.merge(titles, dataset_basics, on="titleId")
title_movies_region.shape
```

```python
title_movies_region.drop(columns=['region','titleType','primaryTitle','originalTitle','startYear','endYear','runtimeMinutes','isAdult'], axis = 1, inplace = True)
title_movies_region
```

```python
title_movies_region_rating=pd.read_csv("ratings.csv")
title_movies_region_rating.rename(columns={'tconst':'titleId'}, inplace=True)
title_movies_region_rating=pd.merge(title_movies_region, title_movies_region_rating, on="titleId")
title_movies_region_rating
```

```python
title_movies_region_rating_principals=pd.read_csv("principals.csv")
title_movies_region_rating_principals.rename(columns={'tconst':'titleId'}, inplace=True)
title_movies_region_rating_principals= title_movies_region_rating_principals[title_movies_region_rating_principals.titleId.isin(title_movies_region_rating['titleId'])]
```

```python
title_movies_region_rating_principals.drop(columns=['characters','job','ordering'], axis = 1, inplace = True)
```

```python
title_movies_region_rating_principals.shape
```

```python
title_movies_region_rating_principals=pd.merge(title_movies_region_rating, title_movies_region_rating_principals, on="titleId")

revenue_ds=pd.read_csv("movies.csv", delimiter=';')
revenue_ds = revenue_ds[["revenue","imdb_id"]]
final_df=pd.merge(title_movies_region_rating_principals,revenue_ds, left_on="titleId",right_on="imdb_id")
final_df=final_df.loc[final_df['revenue'].astype(int)>0]
final_df
```

```python
names = pd.read_csv('name.csv')
title_basics = pd.read_csv('title.basics.csv')
participants=pd.merge(names,final_df , on="nconst")

director_df=participants.loc[participants['category']=='director']
director_average = director_df.groupby('primaryName')['averageRating'].mean()
#director_average.to_csv('directors_ratings.csv')

ddf = director_average.to_frame()
combined_ddf = pd.merge(names,ddf , on="primaryName")

for index, row in combined_ddf.iterrows():
    allTitles = ''
    for title in row["knownForTitles"].split(','):
        curr = title_basics.loc[title_basics['tconst']==title]
        if(len(curr)>0):
            curr = curr.iloc[0]["primaryTitle"]
            allTitles += curr + ', '
    
    combined_ddf.iloc[0]["knownForTitles"] = allTitles
    break
   
#actor_df=participants.loc[(participants['category']=='actor') | (participants['category']=='actress')]
#actor_average = actor_df.groupby(['primaryName','category'])['averageRating'].mean()
#actor_average.to_csv('actor_ratings.csv')
```

```python
final_df['category']
```

```python
final_df=final_df.loc[(final_df['category']).isin (['actor','actress','director','writer'])]
final_df=(final_df[final_df['genres']!="\\N"])
final_df=final_df.drop(columns=['language'])
```

```python
s1 = final_df.genres.str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)
df1 = pd.concat([s1], axis=1, keys=['genres'])
df= final_df.drop(['genres'], axis=1).join(df1).reset_index(drop=True)
df=df.reset_index(drop=True)
```

```python
df = df.drop(['imdb_id'], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(copy=True, feature_range=(0, 1))
columns_to_scale = ['averageRating','numVotes','revenue']
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
#genre=pd.get_dummies(df, columns=[ "genres"], prefix=["gen"])
#genre=genre.drop(columns=['title','nconst','category','titleId'])
df_final=df.drop(columns=['title','nconst','titleId'])
df_final=pd.get_dummies(df_final)
df_final
```

```python
df_final.columns
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
# genre=df[df[df.columns[pd.Series(df.columns).str.startswith('gen')]]==1]
# person_category=df2[df2[df2.columns[pd.Series(df2.columns).str.startswith('per')]]==1]
#titleID=df[df[df.columns[pd.Series(df.columns).str.startswith('ttl')]]==1]

# nameID=df2[df2[df2.columns[pd.Series(df2.columns).str.startswith('nm')]]==1]

feature=df_final.loc[:, (df_final.columns != 'averageRating')]
X_train, X_test, y_train, y_test = train_test_split(feature,
                                                   df_final['averageRating'].astype(float),
                                                     test_size=0.2,
                                                     random_state=11)
```

```python
X= feature
y=df_final['averageRating'].astype(float)
```

```python
plt.figure(figsize=(15,12))
cor = df_final.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
cor_target = abs(cor["averageRating"])
relevant_features = cor_target[cor_target>0.1]
relevant_features
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np

names = X.columns
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
preds = rf.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
accuracy = rf.score(X_test,y_test)
print(accuracy*100,'%')
print("Features sorted by their score:")
important_feature = sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), names), 
             reverse=True)
```

```python
important_feature
```

```python
xg_reg = xgb.XGBRegressor(colsample_bytree=0.7,gamma=0.3,max_depth=4,min_child_weight=4,subsample=0.9)
xg_reg.fit(X_train,y_train)
names = X.columns
preds_xg = xg_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds_xg))
#print("feature importance",xg_reg.feature_importances_)
print("RMSE: %f" % (rmse))
print(r2_score(y_test, preds_xg)) 
print(mean_squared_error(y_test, preds_xg)) 

print("Features sorted by their score:")
important_feature_xg = sorted(zip(map(lambda x: round(x, 4), xg_reg.feature_importances_), names), 
             reverse=True)
```

```python
important_feature_xg
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada_boost_regr=AdaBoostRegressor(base_estimator=None, learning_rate=1.0, loss='linear',
        n_estimators=1000, random_state=101)
ada_boost_regr.fit(X_train,y_train)

preds = ada_boost_regr.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
accuracy = ada_boost_regr.score(X_test,y_test)
print(accuracy*100,'%')

print("Features sorted by their score:")
important_feature_ada = sorted(zip(map(lambda x: round(x, 4), ada_boost_regr.feature_importances_), names), 
             reverse=True)
```
